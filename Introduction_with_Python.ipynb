### **Quiz 2**

Buatkan sebuah function yang akan mempunyai Parameter `name`. Fungsi ini akan mengembalikan nama kalian. Ingat gunakan `return` pada fungsi, jangan `print` langsung di dalam fungsinya. <br>

Ouput dari fungsi ini adalah mengembalikan variabel `name` dalam bentuk list.


**example**
<br>

name = Romi Wahyudi

result = [Romi, Wahyudi]


note. Jika nama kalian hanya satu kata, tambahkan "suka belajar nlp"


kalian bisa menggunakan fungsi `split` string untuk tugas ini

```python
def function(name):
    result = name.split() # split untuk memisahkan string berdasarkan spasi dan membuatnya menjadi list

    if len(result) == 1: # 1 karena merupakan elemen pertama bukan dihitung dari index(0)
        result.append("suka belajar nlp")

    return result
```

```python
# masukkan fungsi pada variabel result
result = function("Kipli")
result
```
